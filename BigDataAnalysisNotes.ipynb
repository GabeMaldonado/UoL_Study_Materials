***TF.IDF*** Terms Frequency times Inverse Document Frequency is the formal measure of how concentrated into relatively few documents are the occurrences of a give word. It is calculated by:

$$TF_{ij} = \frac{f_{ij}}{max_{k} f_{kj}} $$

where $f_{ij}$ = frequency / number of occurrences

$i$ = term word

$j$ = document 

$$IDF_i = log2(N / n_i)$$

Example:

In a repository of ten million documents, what is the IDF for a word that appears in a) 40 documents ; b) 10,000 documents

a) IDF = $log10(10,000,000 / 40) = 5.4$

b) IDF = $log10(10,000,000 / 10,000) = 3$

In a repository with ten million documents and a word *w* appears in 320 of them. In a particular document *d* the MAX number of occurrences of the word is 15. Approsimately, what is the TF.IDF score for *w* if that word appears a) once ; b) five times

IDF = $log10(10,000,000 / 320) 4.5$

a) TF = 1/5 = 0.06
   
   TF.IDF = 0.006 * 4.5 = 0.3

b) TF = 5 / 15 = 0.33

   TF.IDF = 0.33 * 4.5 = 1.48

## Intro to Haddop and MapReduce

MapReduce is a programming framework that can coordinate the pooled resources and organize the computing tasks of a given application in a particular way that it makes easy for us to take the advantage of a large number of machines. In other words, MapReduce is a programming model that structures the computings tasks in such a way that takes the advantage of the given cluster. 

#### Scale-up and Scale-out

Scale-up approach
* adds fewer, more powerful servers capable of handling larger workloads
* Scale-up approaches  might not work beyond a certain point
* As the data grows, the worload will increase beyond the capacity of a scale-up server
* Reaching the capacity of a scale-up server may ultimate trigger a scale-out operation
* A mix of scale-up and scale-out approaches may vcompound the costs and weaknesses of both. 
* Scale-up approaches a rarely used in the big data processing field

Scale-out approach
* adds multiple, smaller servers running in a distributed computing model
* physical storage space and hard drive costs are decreased but data reading rates for the dirvers are not. For example, reading one TB of data from the device takes a long time. On the other hand, reading 10GB of data, each from hundred different drives takes much less times. 
* Scaled-out architectures are the de-facto standard in big data.

#### Data Storage in HDFS

HDFS is a file system designed for storing very large files with streaming data access patterns running on a cluster of commodity hardware. It

* uses *streaming data access pattern* as a data access solution
* is optimized for throughput over latency
* is optimized for workloads that are generally of the write-once and read-many type
* handles node failures

### MapReduce

Why is MapReduce necessary?
* stores data redundantly on multiple nodes
* moves computation close to the data to minimise data movement
* simplified programming model to hide complexity 

What is MapReduce?
* MapReduce prograns are inherently parallel, putting large-scale data analysis into the hands of anyone with enough machines at their disposal
* Hadoop can run MapReduce programs written in various programming languages (e.g. Java, Ruby, python, C++)

A more formal definition:
* **Input:** a set of key-value pairs
* Programmer specifies two methods:
   1. **Map(k, v)**. --> <k', v'>*
      * takes a key-value pair and outputs a set of key-value pairs
      * there is one Map call for every (k, v) pair
   2. **Reduce(k', <v'>*)** --> <k', v"> *   
      * all values *v'* with the same key *k'* are reduced together
      * there is one Reduce function call per unique key *k'*

### Intro to Clustering

Clustering is the process of grouping similar data items together. Clustering algorithms belong to the category of unsupervised machine learning. Unsupervised machine learning algorithms find the pattern on hidden structures in the given data for which no training data is available. Assuming we have a collection of datapoints, that when plotted look like a thick cloud  in an N-dimensional space. Each and every point in the graph represents a high dimensional data point. Waht is the dimension of the data? Roughly soeaking, the dimension of the data is the number of attributes each observation has. For instance, if the cloud of data points represent observations that have the height and weigth of people living in a city as attributes then the dimension of the data is 2. 

**The Problem of Clustering**
Given a set of points, with a notion of distance between points, group the points into
some number of clusters, so that

* members of a cluster are close/similar to each other
* members of different clusters are dissimilar.


Usually:
* points are in a high-dimensional space
* similarity is defined using a distance measure:
* Euclidean, Cosine, Jaccard, edit distance, etc…
  * Example: 
  * Data point 1: Height = 5 ft 5 in, Weight = 68 kg
  * Data point 2: Height = 5 ft 6 in, Weight = 70 kg
  * Data point 3: Height = 6 ft 1 in, Weight = 85 kg

### Euclidean Distance
Euclidean distance between to vectors $[x_1, x_2, ..., x_n ]$ in n-dimensional space is defined as:

$$d([x_1, x_2, ..., x_n], [y_1, y_2, ..., y_n]) = \sqrt{\sum_{i=1}^{n}({x_i}^{2}-{y_i}^{2})^{2}}$$

Euclidean distance in a 2-dimensional space is:

$$d([x_1, x_2], [y_1, y_2]) = \sqrt{(x_1 - y_1)^2 + (x_2 - y2)^2}$$

Assume random points within a bounding box e.g. values between 0 ansd 1 in each dimension.  The range of the distanc measure is $[0, \sqrt{n}]$

**Why is so difficult?**

* Clustering in two dimensions looks easy.
* Clustering small amounts of data looks easy.
* In most cases, looks are not deceiving.
* Many applications involve not 2, but 10 or 10,000 dimensions.
* High-dimensional spaces look different: almost all pairs of points are at about the
same distance.

***The Curse of Dimensionality***
High-dimensional Euclidean spaces have a number of unintuitive properties that are sometimes referred to as the *"curse of dimensionality"*. Non-Euclidena spaces usually have the same anomalies as well. One manifestation of the curse of dimensionality is that *in high dimensions, almost all pairs of points are equally far away from one another*. Another manifestation is that *almost any two vectors are almost orthogonal.*

**Non-Euclidean measures: Cosine and Jaccard**
There are different ways of representing given objects (e.g. CDs or documents) so that we can use a sensible measure for classification.
**Sets as vectors*: measure similarity by the cosine distance.
Think of a point as a vector from the origin (0,0,...,0) to its location. Two points’ vectors make an angle, whose Cosine is the normalised dot- product of the vectors: $d1.d2/|d2||d1|$.
Sets as sets: measure similarity by the Jaccard distance. For the given two sets X and Y, Jaccard distance = $1 - \frac{size(X\cap Y)}{sixe(X \cup Y)}$

#### Methods of clustering

* Hierarchical
   * agglomerative (bottom-up)
      * initially each point is a cluster
      * repeadtedly combine the two *nearest* clusters into one
   * divisive (top-down)
      * start eith the one cluster and recursively split it

   * Point assignment
      * maintain a set of clusters
      * points belong to the *nearest* cluster

#### K-means clustering and k-mean algorithms 

* Assumes Euclidean space/distance

* Start by picking **k**, the number of clusters
* For example, if the given cluster has three points, $(x_1, y_1), (x_2, y_2)$ and $(x_3, y_3)$ the centroid is:

$$(\frac{x_1+x_2+x_3}{3}, \frac{y_1+y_2+y_3}{3})$$

* Initialize clusters by picking one point per cluster
* Example: pick one point at random, then *k-1* other points, each as far away as possible form the previous points

#### Populating Clusters

1. For each point, place it in the cluster whose current centroid it is nearest
2. After all points are assigned, update the locations of teh centroids of the *k* clusters
3. Reassign all points to their closest centroid. Sometimes moves points between clusters
4. Repeat 2 and 3 until convergence. **Convergence** - points don't move betwen clusters and centroids stabilise. 

#### Getting the k right

How to select *k*?
* Try different ks, looking at the change in the average distance to the centroid as k increases
* The average falls rapidly until the right *k* then changes a little
