## Problem Solving Agents

The environment is:

* Observaable - the current state of the world is known

* Discrete - a finite set of action to choose from in any state

* Deterministric - each action has one outcome

Under such assumptions, the solution to any problem is a sequence of actions. 

A search algorithm takes as problem as input and outputs a sequence of actions. 

### Well-defined Problem
A problem can de defined by:

* The initial state the agent starts in
* A description of possible actions available to the agent
* The transition model - a description of what each actions does and the state space - the set of all states reachable from the initial state
* Paths and step costs
* The goal

### Measuring problem-solving performance

The performance of a search algorithm can be evaluated by:

* **Completeness** does the algorithm guarantee to find a solution? (if there is one)
* **Optimality** does the algorithm find an optimal solution?
* **Time complexity** how long does it take to find a solution?
* **Space Complexity** how much memory is needed to perform the search?

### Uninformed/blind search
Agent:

* Only has the problem definition
* Decides on the next action and evaluates if it achieved the goal satate

Blind Search Strategies:
* breadth-first -- 1st- the root node is expanded then all the successors of the root node are expanded then their successors and so on

* uniform-cost --  
* depth-first search strategies - expands the deepest node first in the current frontier of the search tree


### Informed (heuritic) search

Agent has:

* the problem definition
* Problem specific knowldege. Knowledge expressed as:
   * Evaluation function $f(n)$ a cost estimate where $n$ is a node
* Heuristic function $h(n)$, as a component of $f(n)$

### Best-first search
The best-first search strategy, a node is selected for expansion based on $f(n)$ as a likely lead to a solution


### Local Search Algoritms:

* not systematic
* use little memory, usually O(n)
* can find reasonable solutions in large infinite continuous state spaces 

### Genetic Algorithms 
GA are inspired by charles Darwin's theory of natural evolution. Theya re great for finding optimal or good enough solutions in complex situations where most standard approaches may struggle. 

* The fittest individuals are selected for reporduction
* The offspring inherit the characteristics of the parents
* Individuals in a new generation are assessed, the fittests are selected for further reproduction, and so on
* The process converges when there are no improvements in the fitness over the next generations

The idea is that natural selection will find a solution for the problem and it often does but not always. 

GA has three evolutionary processes:

1- Selection
2- Gene crossover
3- Mutation

GA has five components:

1- Initial population
2- Fitness function

### Nondeterministic and partially observable enviroments 
* If the next state of the environment is completely determined by the current state and the action executed by the agent then it is deterministic environment otherwise it is nondeterministic or stochastic environment

* If an agent't sensors give access to the complete state of the environment at any time then the environment is fully observable if not then-- it is partially observable or non-observable environment.

**Agent's percepts**

Percepts provide additional information about the environment:

* if nondeterministic, percepts tell the agent which of the possible outcomes of actions actually occurred
* if partially observable, percepts narrow down the sets of possible states
* Percepts cannot be determined in advance
* Agent's future actions depend on future percepts

A solution to problem is a **contingency plan** -- what to do depending on what percepts are received. 

This is how humans act and interat in the real world. The environment we live in is stochastic. We don't know with certainty what is going to happen tomorrow or in a month. We do not fully understand how many areas work, such as financial markets, or how many students will enroll in the next semester yet we can make short-term and long-term decisions. We are pretty good at it and this is a form of intelligent behavior.

### Optimal strategies for making decisions

In games:

* An agent aims to maximize its utility value by selecting the best possible move that leads to an outcome with the best utility
* Under the assumption that the opponent plays to minimize the utility value

Such search strategy is called minimax

* A minimax search tree is similar to AND-OR tree

MAX-node: OR-node (the player)
MIN-node: AND-mode (the opposition)

### Minimax algorithm modifications

Alpha-beta pruning cuts away brunches of the search
trees that can’t influence the final outcome
* The order of the nodes: If the best options are
considered first, then larger portions of the search
trees can be pruned
* Employing additional knowledge, e.g. remembering the
moves that were found to be the best in the past.
* The best moves are called killer moves, and the
search strategy – the killer move heuristic

The minimax algorithm works with the entire game
tree

* Alpha-beta pruning cuts off unpromising branches
* For real-time decisions, reduction of the search
depth:
* a fixed depth, or
* an iterative deepening depth

## Quantifying Uncertainty

* Uncertainty can be due to various factors:
   * Knowledge gaps
   * Contradictory information from different sources
   * Caustic nature of the environment
* An intelligent agent should be able to made decisions under uncertainty and handle it because:
  * the environment is partially observable
  * the environment is nondeterministic
  * the environment is partially observable and nondeterministic
* An agent needs to:
  * Represent the uncertain knowledge
  * Reason over certain knowledge
For instance, it can be hard o determined the cause of a toothache. A cavity can be a reason for it but it is not for ceertain. There might be other factors casusing pain.

* Logic is not the best option for capturing uncertain knowledge because:
  * We often can't enumerate all possible outcomes, symptoms, causes
  * Our knowledge is limited, uncertain, and it has gaps
  * It can be difficult to do from practical considerations, e.g. too many posibilities to list.
For the toothache example, we can provide a degree of believe:
* 80% change that the toothache is due to a cavity
* P(toothache -> Cavity) = 0.8

 **Probability** provides a way of quantifying the uncertainty. 

 * Preferences expressed as utilities are combined with probabilities in the general theory of rational decisions are called  **Decision Theory**

 * An agent is rational if and only if it chooses the action that yields the highest espected utility averaged over all teh possible outcomes of the action

 * This is the principle of the maximum expected utility(MEU)


$$Decision\ theory = probability\ theory + utility\ theory$$

* Probabilistic assertions are about possible worlds, how probable possible worlds are
* The set of all possible worlds is called **the sample set** denoted by a capital omega Ω
* The possible worlds, denoted as a lower case omega ɷ are mutually exclusive and exhaustive. Example: rolling two dice has 36 possible words (1, 1), (1, 2) ..., (6, 6)

#### Axioms of Probability Theory

* Each possible world has a probability between 0 and 1 

$$0 \leq P (ɷ) \leq 1 $$

* The total probability of all possible worlds is 1:

$$\sum_{w∈Ώ} P (w) = 1$$

Example: rolling two dice. Each world has the probability $P(w) = 1/36$

* Probabilistic assertions  are usually about sets of them-- *events*
* Events are described in AI as propositions in a formal language
* The probability associated with a proposition is defined to be the sum of the probabilities of the wolrds in which it holds. Example: the probability of the event where doubles are rolled up is:

$$P(doubles$= P(1,1) + P(2,2) + P(3,3) + P(4,4) + P(5,5) + P(6,6)$$
$$= 1/36 + 1/36 + 1/36 + 1/36 + 1/36 + 1/36 = 1/6$$

* Probabilites representing the degree of believe in proposition in the absence of any other information or evidence are called ***unconditional or prior probabilities***. Example: P(doubles), P(Total=11)

* Probabilities representing the degreee of believe in proposition after seeing some evidence are called ***conditional or posterior probability*** Example: rolling doubles and the first die shows 5 = P(doubles | Die 1 = 5)

* Variables in probability theory are called **Random Variables** 
* Every random variable has a domain -  the set of posible values it can take on. Example: the domain of the variable Die is {1, 2, 3, 4, 5, 6}. The domain of the variable weather can be {sunny, rainy, cloudy, snowy}.
* Probability of all possible values (probability distribution of the random variable weather):
  * P(weather = sunny)    = 0.6
  * P(weather = rainy)    = 0.1
  * P(weather = cloudy)   = 0.29
  * P(weather = snowy)    = 0.01
  * The total probablities of all values always = 1

* **Independence** -- two events *a* and *b* are independent if:
$$P(a|b) = P(a)$$
  * The fact that *b* has happened, has no effect on *a* happening.
  * Example: *a* = sunny day, *b* = increase in oil price

### What is a probability?

* **Frequentist** (objective) position is that the numbers can come only from experiments. E.g. if we test 100 people and find 20 have cavities then: $P(cavity) = 0.2$

* **Bayesian** (subjective) position is we can nevertheless assign a number based on our knowledge or believes. 

### Baye's Theorem:

$$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$$ where P(B)!= 0

* A and B are events
* P(A|B) is the likelihood of event A occurring given that B is true
* P(B|A) is the likelihood of event B occurring given that A is true
* P(A) and P(B) are the probabilities of observing A and B respectively
* Prior probability P(A) represents what is originally believed before observing B
* Posterior probability P(A|B) takes this new information into account

### Test Driving Development

These tests, intended to initially fail in the absence of any substantial implementation, effectively constitute a specification of the functionality and behavior of the software code which can be tested as it is being developed.
