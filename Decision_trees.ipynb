**Task 1:**
Using what you learnt in the scikit-learn decision tree tutorial, use decision trees for classification on the iris dataset, and for regression on the diabetes dataset (both included in ```sklearn.datasets```). Your code should print the accuracy and the confusion matrix for the classification problem, and mean squared error for the regression. Try comparing the results for different maximum tree depths. 

Note: You should split your data 80% training and 20% for testing.



````python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()

X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# create an instance of the Decision Tree
dt = DecisionTreeClassifier(max_depth=3)

# Fit the DTC
dt.fit(X_train, y_train)

# predict values
y_pred = dt.predict(X_test)

# print valuesrint(y_test)
print(y_test)
print(y_pred)







````

````python
# Evaluate the DTC on the iris dataset

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
````

````python
# Now evaluate a regression on the diabetes dataset 
from sklearn.tree import DecisionTreeRegressor

diabetes = datasets.load_diabetes()
print(diabetes.DESCR)

````

````python
# Load & split data

X2 = diabetes.data
y2 = diabetes.target

# split data

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.3)
````

````python
# create an instance of the DTC with a max_depth of 3
dt2 = DecisionTreeRegressor(max_depth=3)

# fit the classifier
dt2.fit(X2_train, y2_train)

# predict values 
y2_pred = dt2.predict(X2_test)

print(y2_test)
print(y2_pred)

````

````python
# Format plot to enhance ETS graph

from pylab import rcParams

rcParams['figure.figsize'] = 12,8
````

````python
from sklearn import tree
tree.plot_tree(dt2);
````

````python
# plot these outputs using matplotlib

import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

plt.plot(y2_test, color='b', label=True)
plt.plot(y2_pred, color='r', linestyle=':', label='prediction')
plt.legend()

plt.title('R^2 score: {}'.format(r2_score(y2_test, y2_pred)))
plt.show()
````
