## Making sense of data

The starting point of data anlysis is a data table, often called a **data set** which contains measured or collected data values represented as numbers or text. 

A **nominal scale** describes a variable with a limited number of different values that cannot be ordered. For instance, a variable **Industry** would be nominal if it ha categorical values such as *financial*, *engineering*, or *retail*. 

An **ordinal scale**  describes a variable whose values can be ordered or ranked. For example, a scale where the only values are *low*, *medium*, and *high* tells the ranked of the values even though the magnitude of their differences cannot be seen. 

An **interval scale** describes values where the interval between values can be compared. For instance, temperature measurements of 5, 10,15, 20 degrees F.

A **ratio scale**  describes variables where both intervals and ratios of values can be compared. For instance, a bank accuount values of 10, 15, 20 USD, we can clearly see the incremental iterval as well as ratio -- 20 is twice as much as 10.

## Shape

### Skewness 
Quantifies the asymmetry of the distribuition, or lack thereof. It measures how much the distribution leans to the left or the right. 

### Kurtosis
Quantifies the tailedness, the shape of a distribution's tails in relation to its overall shape. It measures the type of peak of the distribution. Variables with pronounced peaks near the mean have a high kurtosis scores while variables with a flat peak have a low kurtosis.


```
# calculate and graph skewness
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# set the random number generator
np.random.seed(6)

s = pd.Series(np.random.normal(0, 1, 300))
ax = s.plot.hist(density=True)

quantiles = np.linspace(stats.norm.ppf(0.001),
                        stats.norm.ppf(0.999), 100)

ax.plot(quantiles, stats.norm.pdf(quantiles))

ax.annotate(f' skew = {s.skew().round(2)}', xy = (-3, 0.35))
ax.set_xlabel('Standard Normal Quantiles')
ax.set_ylabel ('Density')
ax.set_title('Non-skewed  Normally Distributed Random Data ')
plt.show()


```

```
# calculate and graph kurtosis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# set the random number generator
np.random.seed(6)

s = pd.Series(np.random.normal(0, 1, 300))
ax = s.plot.hist(density=True)

quantiles = np.linspace(stats.norm.ppf(0.001),
                        stats.norm.ppf(0.999), 100)

ax.plot(quantiles, stats.norm.pdf(quantiles))

ax.annotate(f' kurtosis = {s.kurtosis().round(2)}', xy = (-3, 0.35))
ax.set_ylabel ('Density')
ax.set_title('High Kurtosis Distribution ')
plt.show()

```

# Design Principles in Data Visualization

*  Who will use the reuslts, why and how?
*  Visualizations help build an argument
*  Essential not to mislead or distort

## General Principles

1 **Show the data** -- representations of the data are the most important thing. Draw the viewer's attention to the data not something else. 
* visual symbols
* categorical labels
* lines
* textures
* axes /scales

2 **Simplify** 
* choose the visual that most effectively communicates the information
* draw it as simple as possible

3 **Reduce clutter**
* avoid visualization details that crowed or obscure the data
* when using grids, ensure they are subtle, e.g. light shad eof grey
* avoid unnecessary axes ticks marks
* avoid redundancy such as including numerical labes when the values are already represented graphically
* avoid purely aesthetic embelishments

4 **Revise**
* in EDA, we work very closely with the data
* "quick and dirty" visualizations can be useful in building up an understanding of the data
* to communicate with data requires clear and polished visuals
* it is critical to iteratively revise a visualization to find the most efficient form
* often involves removing unnecessary details
* always be guided by the message you are trying to convey

5 **Be honest**

* it is a big responsibility
* a lot of trust is placed in the technical ability and integrity as individuals

## Gestalt Laws
1. Proximity -- place symbols and glyphs representing related information close together.
2. Similarity -- when designing a grid layout of a data set, consider coding rows and or columns using low-level visual channels properties, such as color and texture.
3. Connectedness -- to show relationships between entities, consider linking graphical representations of data objects using lines or ribbons of color.
4. Continuity -- one is more likely to construct visual entities out of visual elements that are smooth and continous, rather than ones that contain abrupt changes in direction. 
5. Symmetry -- use symmetry to make pattern comparison easier but be sure that the patterns to be compared are small in terms of visual angle (<1 degree horizontally and <2 degress vertically).
6. Closure --consider putting related information inside a closed contour. A line is adequate for regions having a simple shape. Color or texture can be used to define regions that have more complex shapes.

## Intro to Graph and Networks

* The study of graphs is a key topic in Mathematics
* It is formalyy known as graph theory
* Graphs have significant applications in Computer Science and Data Science
* Graph theory provides the theoretical background for solving various network-related problems
* Graphs are the primary tool whoch can be used to buil and process various types of networks

Graphs consist of two main components:
   * Nodes which represent the actual data/information
   * Edges which provide the connections between the data
Some sources denote nodes as vertices or points and edges are also known as links or lines.
A graph can be a set of ordered pairs where each pair represents two connected nodes. 

Formal definition of a graph:
* Grapgh G comprises of a non-empty finite set of nodes $N(G)$ and a finite set of edges $E(G)$.

Based on this defintions, we can conclude that the graph:
* requires at least one node
* does not have to be connected
* both nodes and edges are finite

Computer Science further develops the concepts of graphs and by utilizing them provide solutions to a range of network-related problems. Some of the more common tasks are:
* Traversing a graph
* Finding optimal (shortest) path

Traversing a graph is visiting all nodes by following the available connection/edges. This is not a trivial task  due to the non-linear structure of the graphs.
The two most common algorithms for traversing graphs are:
* Breadth-firts search (BFS)
* Depth-firts search (DFS)

### Types of Graphs
Some of the most popular graphs include:

* **Undirected graph:** all edges connect the nodes in both directins
* **Directed graph:** the edges connect the nodes in one direction
* **Complete graph:** every node is connected to all other nodes
* **Wighted graph:** every edge has a value assigned to it
* **Tree:** a graph with root node and without cycles
* **Cyclic graph:** a graph with more than one route between nodes

### Applications for Graphs

As a programming tool, graphs have applications in many areas and can be used to model, simulate and visualize numerous different types of networks. Examples include:
* social networks
* computer networks/comms networks
* interconnected web pages
* relational databases
* power grids
* transport networks
* Several optimization problems in Mathematics and Theoretical Computer Science. 

### Properties of Graphs

Graphs have a number of properties/parameters which can be used to describe them. Some of the most frequently used are:

* **Order** which is the number of nodes
* **Size** which is the number of edges
* **Degre of a node** which is the number of edged connecting that node. if equal to $1$ the node is called a *leaf* and it equal to $0$ the node is isolated.

### Directed Graph

* All edges have directions
* The direction is represented with arrows
* Edges can be presented as ordered pairs (A, B)
* Traversig the edges in the directed graph is restricted to one direction only
* e.g. Twitter account $X$ follows $Y$ but $Y$ does not follow $X$


```
# Directed Graph

di_graph = nx.DiGraph()

di_graph.add_node('A')
di_graph.add_node('B')
di_graph.add_node('C')
di_graph.add_node('D')

# Add edges
di_graph.add_edge ('A', 'B')
di_graph.add_edge ('A', 'C')
di_graph.add_edge ('A', 'D')
di_graph.add_edge ('C', 'B')

nx.draw_networkx(di_graph)
plt.show();
```

### Complete Graph

* In this graph, every possible pair of nodes is connected to a unique edge
* As a data structure, is not different from an undirected graph
* Except all possible edges are connected
* Complete graph with $N$ nodes has $N * (N-1) / 2$ edges

### Weighted graph
* For this graph, a value is assigned to every edge
* The value might represent:
  * distance
  * cost
  * or any numerical value which characterizes the transition from one node to another


```
# Undirected Graph

import matplotlib.pyplot as plt
import networkx as nx

wgraph = nx.Graph()

wgraph.add_node('A')
wgraph.add_node('B')
wgraph.add_node('C')
wgraph.add_node('D')

wgraph.add_edge ('A', 'B', weight='10')
wgraph.add_edge ('A', 'C', weight='20')
wgraph.add_edge ('B', 'C', weight='30')
wgraph.add_edge ('A', 'D', weight='40')

weights = nx.get_edge_attributes(wgraph, 'weight')
pos = nx.spring_layout(wgraph)

nx.draw_networkx(wgraph, pos)
nx.draw_networkx_edge_labels(wgraph, pos, edge_labels=weights)
plt.show();
```

### Contour maps (isoline)

An isoline is a line defined by points of equal scalar value in a 2D scalar field


```
delta = 0.025
x = np.arange(-2.0, 3.0, delta)
y = np.arange(-2.0, 3.0, delta)
xx, yy = np.meshgrid(x, y)
z1 = np.exp(-xx**2 - yy**2)
z2 = np.exp(-(xx - 1)**2 - (yy - 1)**2)
zz = (z1 - z2) * 2

# create the plot

fig, ax = plt.subplots()
im= ax.imshow(zz, interpolation='bilinear', origin='lower',
              cmap=plt.cm.gray, extent=(-2, 3, -3, 3))
contours = ax.contour(xx, yy, zz)
ax.clabel(contours, inline=1, fontsize=10, fmt='%1.1f')
plt.show()
```
